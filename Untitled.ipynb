```python
import os
import shutil
import random

base_folder = r'D:\Image analysis folder\......'
source_folder = os.path.join(base_folder, 'Train')
test_folder = os.path.join(base_folder, 'Test')

classes = ['Healthy', 'Purple blotch']
split_ratio = 0.2  # 20% test, 80% train

for cls in classes:
    source_cls_folder = os.path.join(source_folder, cls)
    test_cls_folder = os.path.join(test_folder, cls)
    train_cls_folder = os.path.join(source_folder, cls)  # original Train folder remains for train images
    
    # Make sure Test folders exist
    os.makedirs(test_cls_folder, exist_ok=True)
    
    # List all images in the class folder
    images = [f for f in os.listdir(source_cls_folder) if f.lower().endswith('.jpg')]
    
    # Shuffle images randomly
    random.shuffle(images)
    
    # Number of images to move to test folder
    n_test = int(len(images) * split_ratio)
    
    # Select test images
    test_images = images[:n_test]
    
    # Move test images to Test folder
    for img_name in test_images:
        src_path = os.path.join(source_cls_folder, img_name)
        dst_path = os.path.join(test_cls_folder, img_name)
        shutil.move(src_path, dst_path)

print("Data split completed successfully!")
```

```python
from PIL import Image
import os

# Folder path
folder_path = r'D:\Image analysis folder\Chilli\Net\........'

# Desired crop size
crop_width = 1280
crop_height = 562

# Loop through CC1.jpg to CC301.jpg
for i in range(103, 141):
    filename = f'CH{i}.jpg'
    file_path = os.path.join(folder_path, filename)

    if os.path.isfile(file_path):
        try:
            with Image.open(file_path) as img:
                w, h = img.size

                if w < crop_width or h < crop_height:
                    print(f"Skipping {filename} (image too small: {w}x{h})")
                    continue

                # Calculate crop box (center crop)
                left = (w - crop_width) // 2
                top = (h - crop_height) // 2
                right = left + crop_width
                bottom = top + crop_height

                cropped_img = img.crop((left, top, right, bottom))
                cropped_img.save(file_path)  # Overwrite original
                print(f"Cropped: {filename}")

        except Exception as e:
            print(f"Error processing {filename}: {e}")
    else:
        print(f"File not found: {filename}")
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import math
import matplotlib.patches as patches
import matplotlib as mpl
import numpy as np

# ======================
# CONFIGURATION SETTINGS
# ======================
class Config:
    # Paths
    SOURCE_ROOT = r"D:\Image analysis folder\Reprsentative image sample"
    OUTPUT_FOLDER = r"D:\Image analysis folder"
    
    # Crop folders
    CROP_FOLDERS = [
        "Cauliflower", "Mungbean", "Chilli", "Kidneybean", "LargeCardamom",
        "Maize", "Onion", "Rice", "Sesame", "Strawberry"
    ]
    
    # Visualization parameters
    IMAGE_SIZE = (224, 224)  # Standard size for CNN inputs
    GRID_COLS = 6            # Number of columns in the grid
    DPI = 300                # Output resolution
    MAIN_FONT = 'Arial'      # Font family for all text
    
    # Color scheme
    BORDER_COLOR = '#2a2a2a'  # Dark gray for image borders
    TICK_COLOR = '#555555'    # Medium gray for axis ticks
    BG_COLOR = '#ffffff'      # White background for JPG
    TEXT_COLOR = '#333333'    # Dark gray for text
    
    # Layout tuning
    TITLE_PAD = 8             # Padding above title
    TITLE_FONTSIZE = 7        # Title font size
    TICK_FONTSIZE = 6         # Axis tick font size
    BORDER_WIDTH = 1.5        # Image border width
    
    # Output settings
    OUTPUT_NAME = "Combined_Images_of_all.jpg"

# ======================
# HELPER FUNCTIONS
# ======================
def apply_modern_style():
    """Apply consistent modern styling to matplotlib"""
    mpl.rcParams['font.family'] = Config.MAIN_FONT
    mpl.rcParams['font.size'] = Config.TICK_FONTSIZE
    mpl.rcParams['axes.edgecolor'] = Config.TEXT_COLOR
    mpl.rcParams['axes.labelcolor'] = Config.TEXT_COLOR
    mpl.rcParams['xtick.color'] = Config.TICK_COLOR
    mpl.rcParams['ytick.color'] = Config.TICK_COLOR
    mpl.rcParams['axes.grid'] = False
    mpl.rcParams['axes.facecolor'] = 'white'
    mpl.rcParams['figure.facecolor'] = Config.BG_COLOR

def process_image(img_path):
    """Load and process image with error handling"""
    try:
        img = Image.open(img_path)
        if img.mode != 'RGB':
            img = img.convert('RGB')
        img = img.resize(Config.IMAGE_SIZE)
        return np.array(img)
    except Exception as e:
        print(f"Error processing {img_path}: {str(e)}")
        return np.zeros((*Config.IMAGE_SIZE, 3), dtype=np.uint8)

def create_rounded_border(ax):
    """Create rounded rectangle border for images"""
    border = patches.FancyBboxPatch(
        (0, 0), Config.IMAGE_SIZE[0], Config.IMAGE_SIZE[1],
        boxstyle="round,pad=0.02,rounding_size=0.5",
        linewidth=Config.BORDER_WIDTH,
        edgecolor=Config.BORDER_COLOR,
        facecolor='none',
        zorder=10
    )
    ax.add_patch(border)

# ======================
# MAIN VISUALIZATION
# ======================
def create_image_grid():
    # Collect all images with labels
    image_data = []
    for crop_folder in Config.CROP_FOLDERS:
        crop_path = os.path.join(Config.SOURCE_ROOT, crop_folder)
        if not os.path.isdir(crop_path):
            print(f"Warning: {crop_folder} directory not found")
            continue
            
        images = [f for f in os.listdir(crop_path) 
                 if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        for img in images:
            img_path = os.path.join(crop_path, img)
            abnormality = os.path.splitext(img)[0].replace('_', ' ').title()
            label = f"{crop_folder}\n{abnormality}"
            image_data.append((img_path, label))

    if not image_data:
        print("Error: No images found!")
        return

    # Calculate grid dimensions
    num_images = len(image_data)
    rows = math.ceil(num_images / Config.GRID_COLS)
    
    # Create figure
    fig_width = 7.2  # inches (≈183mm)
    fig_height = rows * 1.4  # Adjusted for tight layout
    fig = plt.figure(figsize=(fig_width, fig_height), dpi=Config.DPI)
    apply_modern_style()
    fig.patch.set_facecolor(Config.BG_COLOR)

    # Plot each image
    for idx, (img_path, label) in enumerate(image_data, 1):
        ax = fig.add_subplot(rows, Config.GRID_COLS, idx)
        img_array = process_image(img_path)
        ax.imshow(img_array)
        create_rounded_border(ax)
        
        ax.set_xticks([0, Config.IMAGE_SIZE[0]//2, Config.IMAGE_SIZE[0]])
        ax.set_yticks([0, Config.IMAGE_SIZE[1]//2, Config.IMAGE_SIZE[1]])
        ax.set_xticklabels(['0', '112', '224'], fontsize=Config.TICK_FONTSIZE)
        ax.set_yticklabels(['0', '112', '224'], fontsize=Config.TICK_FONTSIZE)
        
        ax.tick_params(axis='both', which='both', 
                      length=4, width=0.5, 
                      color=Config.TICK_COLOR,
                      direction='out', 
                      pad=2)
        
        ax.set_title(label, fontsize=Config.TITLE_FONTSIZE, 
                    color=Config.TEXT_COLOR,
                    pad=Config.TITLE_PAD,
                    wrap=True)
        
        for spine in ax.spines.values():
            spine.set_visible(False)

    # Adjust layout
    plt.subplots_adjust(
        wspace=0.15, 
        hspace=0.0,  # No row spacing
        left=0.05, 
        right=0.95,
        top=0.95,
        bottom=0.05
    )
    
    # SAVE FIGURE - WORKAROUND FOR JPG QUALITY ISSUE
    output_path = os.path.join(Config.OUTPUT_FOLDER, Config.OUTPUT_NAME)
    
    # First save as temporary PNG
    temp_path = os.path.join(Config.OUTPUT_FOLDER, "temp_highres.png")
    plt.savefig(temp_path, dpi=Config.DPI, bbox_inches='tight', pad_inches=0.1)
    
    # Convert to JPG using PIL with RGB mode
    try:
        img = Image.open(temp_path)
        img = img.convert('RGB')  # Convert to RGB to avoid RGBA issue
        img.save(output_path, 'JPEG', quality=95)
        print(f"Successfully saved high-quality JPG to: {output_path}")
        if os.path.exists(temp_path):
            os.remove(temp_path)  # Clean up temporary file
    except Exception as e:
        print(f"Error converting to JPG: {str(e)}")
        if os.path.exists(temp_path):
            os.remove(temp_path)
    
    plt.close()

if __name__ == "__main__":
    create_image_grid()    
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import math
import matplotlib.patches as patches

# Paths
source_root = r"D:\Image analysis folder\Reprsentative image sample"

# Crop folders
crop_folders = [
    "Cauliflower", "Mungbean", "Chilli", "Kidneybean", "LargeCardamom",
    "Maize", "Onion", "Rice", "Sesame", "Strawberry"
]

# Collect all images with crop and abnormality labels
image_paths = []
labels = []
for crop_folder in crop_folders:
    crop_path = os.path.join(source_root, crop_folder)
    if not os.path.isdir(crop_path):
        print(f"Error: {crop_folder} not found")
        continue
    images = [f for f in os.listdir(crop_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    for img in images:
        image_paths.append(os.path.join(crop_path, img))
        abnormality = os.path.splitext(img)[0]
        labels.append(f"{crop_folder} - {abnormality}")

# Total number of images
num_images = len(image_paths)
if num_images == 0:
    print("No images found!")
    exit()

# Calculate grid dimensions (5 rows x 6 columns for 29 images)
cols = 6
rows = math.ceil(num_images / cols)

# Create figure (183 mm ≈ 7.2 inches, minimized height)
plt.figure(figsize=(7.2, rows * 1.4), dpi=300)

# Plot each image with label, grid line, and axis labels
for i, (img_path, label) in enumerate(zip(image_paths, labels)):
    try:
        img = Image.open(img_path).resize((224, 224))  # Consistent size for CNN
        ax = plt.subplot(rows, cols, i + 1)
        ax.imshow(img)
        # Add grid line (black border)
        border = patches.Rectangle((0, 0), 224, 224, linewidth=2, edgecolor='black', facecolor='none')
        ax.add_patch(border)
        # Set x and y axes with labels at 0, 100, 200
        ax.set_xticks([0, 100, 200])
        ax.set_yticks([0, 100, 200])
        ax.set_xticklabels(['0', '100', '200'], fontsize=6, family='Arial')
        ax.set_yticklabels(['0', '100', '200'], fontsize=6, family='Arial')
        ax.tick_params(axis='both', length=4, color='black', direction='out')
        ax.set_xlim(0, 224)
        ax.set_ylim(224, 0)  # Invert y-axis to match image orientation
        ax.set_title(label, fontsize=6, family='Arial', pad=4)
    except:
        print(f"Error loading image: {img_path}")

# Minimize row spacing
plt.subplots_adjust(wspace=0.05, hspace=0.0)
plt.tight_layout(pad=0.2, w_pad=0.5, h_pad=0.0)
#plt.savefig(r"D:\Image analysis folder\grid_abnormalities_axis_labels.tiff", dpi=300, format='tiff', bbox_inches='tight')
plt.savefig(r"Combine all image.jpg", dpi=300, format='jpg', bbox_inches='tight')
plt.show()
print("Plot saved as D:\\Image analysis folder\\grid_abnormalities_axis_labels.tiff and .eps")
```
